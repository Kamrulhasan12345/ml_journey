```python
# for each value of year, get the lifeExp column and calculate the mean
# aka split by year, apply the mean function to lifeExp, and combine the results
dat.groupby("year")["lifeExp"].mean()
```

```python
import numpy as np
```

```python
# use agg or aggregate to pass in your own function
dat.groupby("year")["lifeExp"].agg(np.mean)
```

```python
# a groupby object doesn't acutlly do anything until you make a calculation
type(dat.groupby("year"))
```

```python
type(dat.groupby("year")["lifeExp"])
```

```python
# calculate multiple descriptive statistics
dat.groupby(["year", "continent"])[["lifeExp", "gdpPercap", "pop"]].agg([np.mean, np.std])
```

```python
# use the ( ) and split parts into their own line to make code more readable
# and take up less horizontal space
(dat
 .groupby(["year", "continent"])
 [["lifeExp", "gdpPercap", "pop"]]
 .agg([np.mean, np.std])
)
```

```python
# use reset_index to "flatten" your dataframe
(dat
 .groupby(["year", "continent"])
 [["lifeExp", "gdpPercap", "pop"]]
 .agg([np.mean, np.std])
 .reset_index()
)
```
